```python
import pandas as pd
df2 = pd.DataFrame({'A': {0: 'a', 1: 'b',  2: 'c'},
                                  'B': {0:  1,  1: 3,   2: 5},
                                  'C': {0:  2, 1: 4,   2: 6}})
df2
```

```python
pd.melt(df2, id_vars=['A'], value_vars=['B'])
```

```python
pd.melt(df2, id_vars=['A'], value_vars=['B','C'])
```

```python
pd.melt(df2, value_vars=['A','B','C'])
```

```python
df2 = pd.melt(df2, value_vars=['A','B','C']).rename(columns={ 'variable' : 'var','value' : 'val'})
df2
```

```python
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two', 'two'],
                    'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                    'baz': [1, 2, 3, 4, 5, 6],
                    'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df
```

```python
df.pivot(index='foo', columns='bar', values='baz')
```

```python
df.pivot(index='foo', columns='bar', values='baz').reset_index()
```

```python
df.pivot(index='foo', columns='bar', values='baz').reset_index().melt()
```
